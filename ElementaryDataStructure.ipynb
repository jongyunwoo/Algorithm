# ❖ Question 1: Stacks
### a) Implement a stack class in Python with methods for push, pop, and is_empty. Test your implementation with various sequences of push and pop operations.
### b) Write a function that checks if a given string has balanced parentheses using a stack to track the opening and closing symbols.
### c) Modify your stack class to include a method get_top that returns the top element without removing it.

```python
#a
class Stack:

    def __init__(self):
        self.data = []
        self.top = -1

    def size(self):
        return len(self.data)
    
    def isempty(self):
        if self.top == -1:
            return True
        else:
            return False
    
    def push(self, x):
        if self.top == self.size():
            raise OverflowError("overflow")
        else:
            self.top = self.top + 1
            self.data.append(x)
    
    def pop(self):
        if self.isempty():
            raise MemoryError("underflow")
        else:
            self.top = self.top - 1
            return self.data.pop()

# 스택 구현 테스트
stack = Stack()
stack.push(1)
stack.push(2)
print(stack.pop())  # 출력: 2
stack.push(3)
print(stack.pop())  # 출력: 3
print(stack.isempty())  # 출력: False
print(stack.pop())  # 출력: 1
print(stack.isempty())  # 출력: True
```

# ❖ Question 2: Queues
### a) Implement a queue class in Python with methods for enqueue, dequeue, and is_empty. Test your implementation with various sequences of enqueue and dequeue operations.
### b) Write a function that uses a queue to simulate a printer queue. Process print requests in the order they are received.
### c) Modify your queue class to include a method get_front that returns the front element without removing it.

```python
class Queue:

    def __init__(self):
        self.q = []
        self.head = -1
        self.tail = -1

    def size(self):
        return len(self.q)

    def isempty(self):
        return self.size() == 0

    def enqueue(self, x):
        self.q.append(x)
        self.tail += 1

    def dequeue(self):
        if self.isempty():
            raise IndexError("underflow")
        x = self.q.pop(0)
        self.head += 1
        return x

# 테스트 예제
queue = Queue()

# enqueue 테스트
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

# dequeue 테스트
print(queue.dequeue())  # 출력: 1
print(queue.dequeue())  # 출력: 2

# 추가 enqueue 및 dequeue 테스트
queue.enqueue(4)
print(queue.dequeue())  # 출력: 3
print(queue.dequeue())  # 출력: 4

# 큐가 비었는지 확인
print(queue.isempty())  # 출력: True
```

# ❖ Question 3: Linked Lists
### a) Implement a linked list in Python. Include methods for insert, prepend, search, and delete. Test these methods on various scenarios.
### b) Create a playlist for a music player. Each node in the linked list will represent a song in the playlist, and you can use the linked list to manage the songs' order.

```python
#a
class Node:
    def __init__(self, item):
        self.data = item
        self.prev = None
        self.next = None

class DoubleLinkedList:
    
    def __init__(self):
        self.nodeCount = 0
        self.head = Node(None)
        self.tail = Node(None)
        self.head.prev = None
        self.head.next = self.tail
        self.tail.prev = self.head
        self.tail.next = None

    def insert(self, prev, newnode):
        next = prev.next
        newnode.prev = prev
        newnode.next = next
        prev.next = newnode
        next.prev = newnode
        self.nodeCount += 1
        return True
    
    def prepend(self, newnode):
        next = self.head.next
        self.head.next =  newnode
        newnode.prev = self.head
        newnode.next = next
        next.prev = newnode
        self.nodeCount += 1
        return True
    
    def search(self, node):
        x = self.head.next
        while x != self.tail:
            if x.data == node:
                return x
            x = x.next
        return None

    def delete(self, node):
        if self.nodeCount == 0:
            raise IndexError("underflow")
        node.prev.next = node.next
        node.next.prev = node.prev
        node.prev = None
        node.next = None
        self.nodeCount -= 1
        return node
    
    def traverse(self):
        result = []
        current = self.head.next
        while current != self.tail:
            result.append(current.data)
            current = current.next
        return result
    
# 테스트 예제
dll = DoubleLinkedList()

# 노드 추가
node1 = Node(1)
dll.prepend(node1)

node2 = Node(2)
dll.insert(node1, node2)  # node2를 node1 다음에 삽입

node3 = Node(3)
dll.insert(node2, node3)  # node3을 리스트 맨 앞에 삽입

# 현재 리스트 상태 확인
print(dll.traverse())  # 출력: [3, 1, 2]

# 노드 검색
searched_node = dll.search(1)
print(searched_node.data)  # 출력: 1

# 노드 삭제
dll.delete(node2)
print(dll.traverse())  # 출력: [3, 1]

# 노드 삭제 후 상태 확인
dll.delete(node1)
print(dll.traverse())  # 출력: [3]

# 노드 삭제 후 상태 확인
dll.delete(node3)
print(dll.traverse())  # 출력: []
```

```python
#b
class SongNode:
    def __init__(self, title):
        self.title = title
        self.prev = None
        self.next = None

class Playlist:
    def __init__(self):
        self.head = SongNode(None)  # 더미 노드
        self.tail = SongNode(None)  # 더미 노드
        self.head.next = self.tail
        self.tail.prev = self.head
        self.current = None
        self.nodeCount = 0

    def add_song(self, title):
        new_song = SongNode(title)
        prev = self.tail.prev
        prev.next = new_song
        new_song.prev = prev
        new_song.next = self.tail
        self.tail.prev = new_song
        self.nodeCount += 1
        if self.current is None:  # 첫 노래를 추가할 때 현재 노래로 설정
            self.current = new_song

    def get_current_song(self):
        if self.current is None:
            return "No song is currently playing"
        return f"Currently playing: {self.current.title}"

    def next_song(self):
        if self.current is None or self.current.next == self.tail:
            return "No next song available"
        self.current = self.current.next
        return f"Playing next song: {self.current.title}"

    def previous_song(self):
        if self.current is None or self.current.prev == self.head:
            return "No previous song available"
        self.current = self.current.prev
        return f"Playing previous song: {self.current.title}"

    def show_playlist(self):
        songs = []
        current = self.head.next
        while current != self.tail:
            songs.append(current.title)
            current = current.next
        return "Playlist: " + " -> ".join(songs)

# 테스트 예제
playlist = Playlist()

# 노래 추가
playlist.add_song("Song 1")
playlist.add_song("Song 2")
playlist.add_song("Song 3")

# 현재 재생 목록 확인
print(playlist.show_playlist())  # 출력: Playlist: Song 1 -> Song 2 -> Song 3

# 현재 노래 확인
print(playlist.get_current_song())  # 출력: Currently playing: Song 1

# 다음 노래로 이동
print(playlist.next_song())  # 출력: Playing next song: Song 2

# 또 다음 노래로 이동
print(playlist.next_song())  # 출력: Playing next song: Song 3

# 다음 노래가 없을 때
print(playlist.next_song())  # 출력: No next song available

# 이전 노래로 이동
print(playlist.previous_song())  # 출력: Playing previous song: Song 2

# 현재 노래 확인
print(playlist.get_current_song())  # 출력: Currently playing: Song 2
```
